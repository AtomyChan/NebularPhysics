Question: What's the ratio of $M_{Co}$ to $M_{Fe}$ if both are product of $^{56}Ni$ decay?

In $^{56}\rm Ni \rightarrow ^{56}Co \rightarrow ^{56}Fe$ process, the numbers of $^{56}$Ni, $^{56}$Co, and $^{56}$Fe nuclides are controlled by a simple set of differential equations:

$\frac{d N_{\mathrm{Ni}}}{d t}=-\frac{N_{\mathrm{Ni}}}{\tau_{\mathrm{Ni}}}$

$\frac{d N_{\mathrm{Co}}}{d t}=\frac{N_{\mathrm{Ni}}}{\tau_{\mathrm{Ni}}}-\frac{N_{\mathrm{Co}}}{\tau_{\mathrm{Co}}}$

$\frac{d N_{\mathrm{Fe}}}{d t}=\frac{N_{\mathrm{Co}}}{\tau_{\mathrm{Co}}}$

with the initial conditions


$N_{\mathrm{Ni}}=N_{\mathrm{Ni} 0}$,  $N_{\mathrm{Co}}=0$, $N_{\mathrm{Fe}}=0$, at t=0

Solving above equations, one gets


\begin{array}{l}
{N_{\mathrm{Ni}}=N_{\mathrm{Ni0}} e^{-(t / \tau_{\mathrm{Ni}})}} \\
{N_{\mathrm{Co}}=N_{\mathrm{Ni0}} \frac{\tau_{\mathrm{Co}}}{\tau_{\mathrm{Co}}-\tau_{\mathrm{Ni}}}\left(e^{-\left(t / \tau_{\mathrm{Co}}\right)}-e^{-\left(t / \tau_{\mathrm{Ni}}\right)}\right)} \\
{N_{\mathrm{Fe}}=N_{\mathrm{Ni0}}\left(1+\frac{\tau_{\mathrm{Ni}}}{\tau_{\mathrm{Co}}-\tau_{\mathrm{Ni}}} e^{-\left(t / \tau_{\mathrm{Ni}}\right)}-\frac{\tau_{\mathrm{Co}}}{\tau_{\mathrm{Co}}-\tau_{\mathrm{Ni}}} e^{-\left(t / \tau_{\mathrm{Co}}\right)}\right)}
\end{array}

```python
import numpy as np
%matplotlib notebook
import matplotlib.pylab as plt

T12Ni = 6.075
T12Co = 77.2
tauNi = T12Ni/np.log(2)
tauCo = T12Co/np.log(2)

ts = np.linspace(210, 500, 100)

MCo2MFe = tauCo/(tauCo-tauNi)*(np.exp(-ts/tauCo)-np.exp(-ts/tauNi))/(1+tauNi/(tauCo-tauNi)*np.exp(-ts/tauNi)-tauCo/(tauCo-tauNi)*np.exp(-ts/tauCo))
```

```python
plt.semilogy(ts, MCo2MFe)
plt.show()
```
